```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
a_dataframe = pd.read_csv("out1.txt", delimiter='\t')
x = a_dataframe['t']
y1 = a_dataframe['x']
y2 = a_dataframe['x_dot']
y3 = a_dataframe['theta']
y4 = a_dataframe['theta_dot']
plt.figure(figsize=(12, 7))

plt.subplot(2, 2, 1)
plt.title("x")
plt.plot(x, y1)
plt.subplot(2, 2, 2)
plt.title("x_dot")
plt.plot(x, y2)
plt.subplot(2, 2, 3)
plt.title("theta")
plt.plot(x, y3 / math.pi * 180)
plt.subplot(2, 2, 4)
plt.title("theta_dot")
plt.plot(x, y4)
plt.show()
```
